## Prepare the Data before applying Facebook prophet tool

```python
df
```

```python
# Selecting only date and average price columns
prophet_df = df[['Date','AveragePrice']]
```

```python
prophet_df
```

```python
# Renaming date and average price columns
prophet_df = prophet_df.rename(columns = {'Date':'ds', 'AveragePrice':'y'})
```

```python
prophet_df
```

## Develop model & Make predictions

```python
p = Prophet()
p.fit(prophet_df)
```

```python
# Forcasting into the future
future = p.make_future_dataframe(periods = 365)
forecast = p.predict(future)
```

```python
forecast
```

```python
plt.figure(figsize=(10,10))
figure = p.plot(forecast,xlabel = 'Date', ylabel = 'Price')
```

```python
figure2 = p.plot_components(forecast)
```

## Develop Model & Make predictions
